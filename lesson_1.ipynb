```python
import pandas as pd
from random import randint
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Генерация случайной последовательности
random_numbers = [randint(1, 20) for _ in range(10)]
# Создаем ось X (номера элементов)
x = np.arange(len(random_numbers)).reshape(-1, 1)
```

```python
# Создаем DataFrame
df = pd.DataFrame({'Value': random_numbers})
```

```python
# Строим линейный тренд с помощью линейной регрессии
model = LinearRegression()
# находим оптимальные коэффициенты (веса) модели, которые минимизируют ошибку предсказания
model.fit(x, random_numbers)
trend = model.predict(x)
```

```python
# Определяем характер тренда
slope = model.coef_[0]
if slope > 0:
    trend_text = 'Линейный тренд ВОЗРАСТАЕТ'
elif slope < 0:
    trend_text = "Линейный тренд УБЫВАЕТ"
else:
    trend_text = "Линейный тренд БЕЗ ИЗМЕНЕНИЙ"
```

```python
# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(x, random_numbers, 'bo-')
plt.plot(x, trend, 'r--')
plt.title(f'Случайная последовательность\n{trend_text}')
plt.xlabel('Индекс элемента')
plt.ylabel('Значение')
plt.grid(True)
plt.show()
```

```python
# Вывод результата
print(f"Угловой коэффициент (slope): {slope:.2}")
print(trend_text)
```
