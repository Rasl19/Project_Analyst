```python
import random
```

```python
import pandas as pd
df = pd.read_csv('customer_segmentation_mixed.csv')
```

```python
# 1. Возрастные группы
def age_group(age):
    if age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 50:
        return '36-50'
    else:
        return '50+'

df['age_group'] = df['Age'].apply(age_group)
```

```python
# 2. Категория трат (Spending Category)
def spending_category(score):
    if score >= 80:
        return 'VIP Spender'
    elif score >= 60:
        return 'High Spender'
    elif score >= 40:
        return 'Moderate Spender'
    else:
        return 'Low Spender'

df['spending_category'] = df['Spending_Score'].apply(spending_category)
```

```python
# 3. Уровень дохода (Income Level)
def income_level(income):
    if income >= 80000:
        return 'High (>80k)'
    elif income >= 50000:
        return 'Medium (50-80k)'
    else:
        return 'Low (<50k)'

df['income_level'] = df['Annual_Income'].apply(income_level)
```

```python
# 4. Посещаемость (Visit Frequency)
def visit_frequency(visits):
    if visits >= 10:
        return 'Very High (10+)'
    elif visits >= 7:
        return 'High (7-9)'
    elif visits >= 4:
        return 'Medium (4-6)'
    else:
        return 'Low (1-3)'

df['visit_frequency'] = df['Monthly_Visits'].apply(visit_frequency)
```

```python
# 5. Добавление затрат в месяц
def cost(value):
    Annual_Income_month = int(value*0.7)
    return random.randint(100, Annual_Income_month)

df['cost_year'] = df['Annual_Income'].apply(cost)
```

```python
df.to_csv('customer_behavior_processed.csv', index=False)
```
